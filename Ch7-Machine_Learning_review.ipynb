## 혼동행렬



- TP : 양성이고 실제로 양성이라서 정답을 맞춘 경우 
- TN : 음성(Negative)이라고 대답하였는데 실제로 음성이라서 정답을 맞춘 경우입니다.
- FP : 음성이라고 대답하였는데, 음성이라서 정답을 틀린 경우
- FN : 음성이라고 대답하였는데 양성이라서 정답을 틀린 경우가 됩니다.

### 정밀도

정밀도 = TP/(TP+FP)

### 재현율

재현율 = TP/(TP+FN)
